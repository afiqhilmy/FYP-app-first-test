```python
import numpy as np
np.random.seed(2111)

def generate_adt(population):
    if population < 25:
        return np.random.randint(8000, 15001)
    elif 25 <= population < 60:
        return np.random.randint(15000, 30001)
    elif 60 <= population < 120:
        return np.random.randint(30000, 50001)
    else:
        return np.random.randint(50000, 80001)

final_merged_df["ADT"] = final_merged_df["population"].apply(generate_adt)
```

# **OPTIMIZING EV CHARGING SCHEDULING (OBJ 3)**

## 1. Peak hour data generation

```python
URBAN_KEYWORDS = [
    "city", "mall", "bandar", "pusat", "square", "komersial",
    "commercial", "downtown", "indera mahkota", "kuantan city"
]
```

```python
def is_urban(address):
    address = address.lower()
    return any(keyword in address for keyword in URBAN_KEYWORDS)
```

```python
# Peak hour flag generator
# -----------------------------
def peak_hour(row):
    urban = is_urban(row["Station Address"])
    pop = row["population"]
    adt = row["ADT"]  # or use real ADT if available

    if urban and (pop >= 60 or adt >= 30000):
        return "Yes"
    else:
        return "No"
```

```python
final_data["Peak_Hour"] = final_data.apply(peak_hour, axis=1)
final_data
```

```python
import numpy as np

final_merged_df["queue_level"] = pd.qcut(
    final_merged_df["predicted_demand"],
    q=3,
    labels=["Low", "Medium", "High"]
)
```

```python
def scheduling_decision(row):
    if row["queue_level"] == "High":
        return "Delay AC, Prioritize DC, Off-peak only"
    elif row["queue_level"] == "Medium":
        return "Limited AC charging"
    else:
        return "Normal operation"

final_merged_df["scheduling_action"] = final_merged_df.apply(scheduling_decision, axis=1)
```

## 3. MILP method

```python
from pulp import LpProblem, LpMinimize, LpVariable

model = LpProblem("EV_Charging_Scheduling", LpMinimize)

stations = final_merged_df.index.tolist()
time_slots = ["peak", "off_peak"]

x = LpVariable.dicts("x", (stations, time_slots), cat="Binary")

# Objective
model += sum(
    final_merged_df.loc[s, "predicted_demand"] * x[s][t]
    for s in stations
    for t in time_slots
)
```

```python
from pulp import (
    LpProblem,
    LpMinimize,
    LpVariable,
    lpSum,
    LpBinary
)

model_milp = LpProblem("EV_Charging_Scheduling", LpMinimize)
```

```python
stations = final_merged_df.index.tolist()

x = LpVariable.dicts(
    "schedule",
    (stations, time_slots),
    cat=LpBinary
)
```

```python
model_milp += lpSum(
    final_merged_df.loc[s, "predicted_demand"] * x[s][t]
    for s in stations
    for t in time_slots
)
```

```python
for s in stations:
    for t in time_slots:
        model_milp += (
            final_merged_df.loc[s, "predicted_demand"] * x[s][t]
            <= final_merged_df.loc[s, "Total Station Capacity (KW)"]
        )

GRID_LIMIT = final_merged_df["Total Station Capacity (KW)"].sum() * 0.6

for t in time_slots:
    model_milp += lpSum(
        final_merged_df.loc[s, "predicted_demand"] * x[s][t]
        for s in stations
    ) <= GRID_LIMIT

for s in stations:
    model_milp += lpSum(
        x[s][t] for t in time_slots
    ) >= 1
```

```python
model_milp.solve()
```

```python
schedule_output = []

for s in stations:
    for t in time_slots:
        if x[s][t].value() == 1:
            schedule_output.append({
                "Station_ID": s,
                "Time_Slot": t
            })

schedule_df = pd.DataFrame(schedule_output)
```

```python
final_merged_df["scheduled_peak"] = [
    int(x[s]["peak"].value()) for s in stations
]

final_merged_df["scheduled_off_peak"] = [
    int(x[s]["off_peak"].value()) for s in stations
]
```

```python
def interpret(row):
    if row["scheduled_peak"] == 0 and row["scheduled_off_peak"] == 1:
        return "Off-peak only (Load shifting)"
    elif row["scheduled_peak"] == 1 and row["scheduled_off_peak"] == 1:
        return "Normal operation"
    else:
        return "Restricted operation"

final_merged_df["scheduling_decision"] = final_merged_df.apply(interpret, axis=1)
```

## 4. Evaluation

```python
# Compute average-to-peak demand ratio per time slot
evaluation = {}
for t in time_slots:
    total_demand = sum(final_merged_df.loc[s, "predicted_demand"] * x[s][t].value() for s in stations)
    peak_demand = final_merged_df["predicted_demand"].max()
    evaluation[f"{t}_avg_to_peak_ratio"] = total_demand / peak_demand

# Compute % of stations meeting demand
satisfied_stations = len(schedule_df['Station_ID'].unique())
num_stations = len(stations)
evaluation["station_satisfaction"] = satisfied_stations / num_stations * 100

print("\nMILP Evaluation Metrics:")
for k, v in evaluation.items():
    print(f"{k}: {v:.2f}")
```
