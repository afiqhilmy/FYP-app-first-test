```python
import re
import pandas as pd

def extract_ac_capacity(text):
    if pd.isna(text):
        return 0
    ac_total_capacity = 0
    # Regex to find (power) followed by 'KW AC', 'KWAC', 'AC' AND (number of units)
    pattern = r'(\d+\.?\d*)\s*KW?\s*AC(?::\s*(\d+)\s*UNIT(?:S)?)?'
    matches = re.findall(pattern, str(text), re.IGNORECASE)
    for match in matches:
        power = float(match[0])
        units_str = match[1] # Number of units, can be None if not found
        units = int(units_str) if units_str else 1 # Default to 1 unit if not specified
        ac_total_capacity += power * units
    return int(ac_total_capacity)

def extract_dc_capacity(text):
    if pd.isna(text):
        return 0
    dc_total_capacity = 0
    # Regex to find (power) followed by 'KW DC', 'KWDC', 'DC' AND (number of units)
    pattern = r'(\d+\.?\d*)\s*KW?\s*DC(?::\s*(\d+)\s*UNIT(?:S)?)?'
    matches = re.findall(pattern, str(text), re.IGNORECASE)
    for match in matches:
        power = float(match[0])
        units_str = match[1] # Number of units, can be None if not found
        units = int(units_str) if units_str else 1 # Default to 1 unit if not specified
        dc_total_capacity += power * units
    return int(dc_total_capacity)
```

```python
import re

def extract_postcode(address):
    # Malaysian postcodes are typically 5 digits
    match = re.search(r'\b(\d{5})\b', address)
    if match:
        return match.group(1)
    return None # Return None if no postcode is found

stations_new['Postcode'] = stations_new['Station Address'].apply(extract_postcode)
stations_new.head()
```

**Create new column average charging hours per day**

```python
ev = pd.read_csv('kuantan_ev_stations.csv')
ev.head()
```

```python
import numpy as np

np.random.seed(1203)

# Average AC charging hours per day (4–8 hours)
ev['Avg_AC_Hours'] = np.where(
    ev['AC_Capacity_kW'] > 0,
    np.random.uniform(8,12, size=len(ev)),
    0
)

# Average DC charging hours per day (3–6 hours)
ev['Avg_DC_Hours'] = np.where(
    ev['DC_Capacity_kW'] > 0,
    np.random.uniform(2, 4, size=len(ev)),
    0
)

# Optional: round to 2 decimal places
ev['Avg_AC_Hours'] = ev['Avg_AC_Hours'].round(2)
ev['Avg_DC_Hours'] = ev['Avg_DC_Hours'].round(2)

# Drop the specified columns
ev = ev.drop(columns=['Avg_AC_Charging_Hours_Per_Day', 'Avg_DC_Charging_Hours_Per_Day'], errors='ignore')

ev.head(2)
```

**Calculating cost and revenue**

7kWh/22kWh AC Charging: RM0.55/kWh

180kWh DC Charging: RM1.00/kWh

350kWh DC Charging: RM1.20/kWh

source: https://soyacincau.com/2022/12/02/ev-dc-fast-charging-in-malaysia-gentari-kwh-vs-time-based-charging-which-is-cheaper/#:~:text=350kWh%20DC%20Charging:%20RM1.&text=The%20pricing%20for%20DC%20fast,to%20charge%20at%20these%20stations.

```python
# Define tariff rates
ac_rate = 0.55
dc_rate_180_kwh = 1.00 # For 180kWh DC Charging
dc_rate_350_kwh = 1.20 # For 350kWh DC Charging

# Calculate AC Revenue
# Using 'Avg_AC_Hours' from the ev DataFrame
ev['AC_revenue (RM)'] = (ev['AC_Capacity_kW'] * ev['Avg_AC_Hours'] * ac_rate).round(2)

# Calculate DC Revenue based on tiers
def calculate_dc_revenue_tiered(row):
    total_dc_capacity = row['DC_Capacity_kW']
    avg_dc_charging_hours = row['Avg_DC_Hours'] # Use Avg_DC_Hours

    if total_dc_capacity == 0:
        return 0.0
    # Apply RM1.00/kWh for DC capacity up to 180kW
    elif total_dc_capacity <= 180:
        return total_dc_capacity * avg_dc_charging_hours * dc_rate_180_kwh
    # Apply RM1.20/kWh for DC capacity greater than 180kW
    else:
        return total_dc_capacity * avg_dc_charging_hours * dc_rate_350_kwh

ev['DC_revenue (RM)'] = ev.apply(calculate_dc_revenue_tiered, axis=1).round(2)

# Display the relevant columns to verify the new calculations
ev[['Station Address', 'AC_Capacity_kW', 'DC_Capacity_kW', 'Avg_AC_Hours', 'Avg_DC_Hours', 'AC_revenue (RM)', 'DC_revenue (RM)']].head()
```

```python
ev['Total_revenue (RM)'] = ev['AC_revenue (RM)'] + ev['DC_revenue (RM)']
ev[['Station Address', 'AC_revenue (RM)', 'DC_revenue (RM)', 'Total_revenue (RM)']].head()
```

```python
import folium

# Calculate the mean latitude and longitude to center the map
mean_lat = ev['Latitude'].mean()
mean_lon = ev['Longitude'].mean()

# Create a Folium map centered at Kuantan
kuantan_map = folium.Map(location=[mean_lat, mean_lon], zoom_start=12)

# Define revenue thresholds based on percentiles
low_threshold = ev['Total_revenue (RM)'].quantile(0.33)
medium_threshold = ev['Total_revenue (RM)'].quantile(0.66)

# Function to determine marker color based on total revenue
def get_revenue_color(revenue):
    if revenue >= medium_threshold:
        return 'green'  # High revenue
    elif revenue >= low_threshold:
        return 'blue'   # Medium revenue
    else:
        return 'red'    # Low revenue

# Add markers for each charging station
for index, row in ev.iterrows():
    if pd.notnull(row['Latitude']) and pd.notnull(row['Longitude']):
        revenue_color = get_revenue_color(row['Total_revenue (RM)'])
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=f"<b>Address:</b> {row['Station Address']}<br><b>Total Chargers:</b> {row['Total Number of Chargers']}<br><b>DC Bays:</b> {row['DC Charger Bays']}<br><b>AC Bays:</b> {row['AC Charger Bays']}<br><b>Total Revenue:</b> RM{row['Total_revenue (RM)']:.2f}",
            icon=folium.Icon(color=revenue_color, icon='charging-station', prefix='fa')
        ).add_to(kuantan_map)

# Display the map
kuantan_map
```
